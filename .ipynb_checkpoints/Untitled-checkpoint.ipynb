```python
df = df[["questions_fixed", "incorrectAnswers", "correctAnswers"]]
```

```python
def format_question_and_answers(row):
    question = row['questions_fixed']
    
    formatted_text = f"\n\nQuestion:\n{question}\n\n"
    try:
        correct_answers = row['correctAnswers'][0]
        formatted_text+=f"Answers:\n\n"
        for answer in row['correctAnswers']:
            formatted_text += f"- [x] {answer}\n"
    except:
        pass
    try:
        incorrect_answers = row['incorrectAnswers'][0]
        for answer in row['incorrectAnswers']:
            formatted_text += f"- [ ] {answer}\n"
    except:
        pass
    
    return formatted_text

# Apply the formatting function to each row
df['formatted_text'] = df.apply(format_question_and_answers, axis=1)
```
